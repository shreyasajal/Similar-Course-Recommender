```python
import nltk
import string
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
keywords=list()
lines=data['course_search'].values.tolist()
for line in lines:
  #tokenize the text
  tokens=word_tokenize(line)
  tokens=[w.lower()for w in tokens]
  #remove punctuations
  table=str.maketrans('','',string.punctuation)
  stripped=[w.translate(table)for w in tokens]
  #remove non-alphabetical characters
  words=[word for word in stripped if word.isalpha()]
  from nltk.corpus import stopwords
  stop_words = set(stopwords.words('english'))
  words = [w for w in words if not w in stop_words]
  keywords.append(words)
```

```python
data['course_search']=pd.Series(np.array(keywords))
data['course_search']
```

```python
def apply_filter(id):
    #filter
    #duration_filter
    if id in first_quarter_df['course_id'].values.tolist():
        first_quarter_df.reset_index(drop=True,inplace=True)
        df2=first_quarter_df
    elif id in second_quarter_df['course_id'].values.tolist():
        second_quarter_df.reset_index(drop=True,inplace=True)
        df2=second_quarter_df
    elif id in third_quarter_df['course_id'].values.tolist():
        third_quarter_df.reset_index(drop=True,inplace=True)
        df2=third_quarter_df
    elif id in fourth_quarter_df['course_id'].values.tolist():
        fourth_quarter_df.reset_index(drop=True,inplace=True)
        df2=fourth_quarter_df
    elif id in fifth_quarter_df['course_id'].values.tolist():
        fifth_quarter_df.reset_index(drop=True,inplace=True)
        df2=fifth_quarter_df
    #field_filter    
    idx = df2.index[df2['course_id']==id]
    if df2['course_field'][idx].values[0]=='-':
        df3=df2
    else:
        df3=df2.loc[df2['course_field']==df2['course_field'][idx].values[0]]
        df3.reset_index(drop=True,inplace=True)
    
    #audit_filter
    idx = df3.index[df3['course_id']==id]

    df4=df3.loc[df3['course_audit']==df3['course_audit'][idx].values[0]]
    df4.reset_index(drop=True,inplace=True)
    
    #certi_filter
    idx = df4.index[df4['course_id']==id]

    df5=df4.loc[df4['course_certi']==df4['course_certi'][idx].values[0]]
    df5.reset_index(drop=True,inplace=True)
    return df5
```
